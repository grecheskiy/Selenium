```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('house-prices/test.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Проверка того, в каких столбцах отсутствуют значения
df.isnull().sum()
```

```python
# Заполнение пропущенных значений медианным значением столбца
median_lfage= df['LotFrontage'].median()
df['LotFrontage'] = df['LotFrontage'].fillna(median_lfage)
```

```python
df.info()
```

```python
# Заполнение пропущенных значений в столбце "Embarked" наиболее распространенным портом
mode_saletype = df['SaleType'].mode()[0]
df['SaleType'] = df['SaleType'].fillna(mode_saletype)
mode_msz = df['MSZoning'].mode()[0]
df['MSZoning'] = df['MSZoning'].fillna(mode_msz)
mode_allay = df['Alley'].mode()[0]
df['Alley'] = df['Alley'].fillna(mode_msz)
mode_utilites = df['Utilities'].mode()[0]
df['Utilities'] = df['Utilities'].fillna(mode_utilites)
```

```python
df.info()
```

```python
duplicates = df.duplicated()
duplicates
```

```python
duplicates_df = df[duplicates]
duplicates_df
```

```python
df = df.drop_duplicates()
```

```python
df.head()
```

```python
# Выполнение label encoding столбца "Embarked"
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['LandContour'] = le.fit_transform(df['LandContour'].astype(str))
```

```python
df.head()
```

```python
# Вычисление Z-score для столбца "Fare"
from scipy import stats

z = np.abs(stats.zscore(df['LotFrontage']))

# Установка порогового значения Z-score
threshold = 3

# Выявление выбросов на основе Z-score
outliers = df['LotFrontage'][z > threshold]
```

```python
df['LotFrontage'].mean()
```

```python
outliers
```

```python
# Замена выбрасов медианным значением
df.loc[z > threshold, 'LotFrontage'] = df['LotFrontage'].median()
```

```python
df['LotFrontage'].mean()
```

```python
# Винсоризация столбца "Fare"
from scipy.stats.mstats import winsorize

df['LotFrontage'] = winsorize(df['LotFrontage'], limits=[0.05, 0.05])
```

```python
df['LotFrontage'].mean()
```

```python
# Трансформация столбца "Fare" с помощью логарифмической функции
df['LotFrontage'] = np.log1p(df['LotFrontage'])
```

```python
df['LotFrontage'].mean()
```
